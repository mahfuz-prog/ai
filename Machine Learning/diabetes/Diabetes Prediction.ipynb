### Importing libraries

```python
import pandas as pd
import numpy as np
```

### loading dataset

```python
df = pd.read_csv('/content/diabetes.csv')
X = df.iloc[:, :-1].values
y = df.iloc[:, -1:].values.ravel()
```

### split the dataset into train & test

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=.3, random_state=1)
```

### feature scaling

```python
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
X_train = sc.fit_transform(X_train)
X_test = sc.fit_transform(X_test)
```

### Machine learning model

```python
from sklearn.svm import SVC
classifier = SVC(kernel='rbf')
classifier.fit(X_train, y_train)
```

### Actual result vs predicted result of test set

```python
ts = pd.DataFrame(data=y_test, columns=['exact result'])
ts['predicted result'] = classifier.predict(X_test)
ts.tail()
```

### new prediction

```python
input = np.array([[1, 103, 80, 11, 82, 19.4, 0.491, 22]])
if classifier.predict(sc.fit_transform(input))[0] == 0:
    print('diabetes detected')
else:
    print('diabetes not detected')
```

### confusion matrix & accuracy

```python
from sklearn.metrics import confusion_matrix, accuracy_score
y_pred = classifier.predict(X_test)
cm = confusion_matrix(y_test, y_pred)
print('Confusion matrix: ')
print(cm)
print('Accuracy: ', accuracy_score(y_test, y_pred)*100, '%')
```

### K-Fold cross validation score & standard deviation

```python
from sklearn.model_selection import cross_val_score
accuracies = cross_val_score(estimator = classifier, X = X_train, y = y_train, cv = 10)
print("Accuracy: {:.2f} %".format(accuracies.mean()*100))
print("Standard Deviation: {:.2f} %".format(accuracies.std()*100))
```
