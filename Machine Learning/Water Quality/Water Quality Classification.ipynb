```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('/content/water_potability.csv')
y = df.Potability
df.drop(columns='Potability', inplace=True)
```

```python
from sklearn.impute import SimpleImputer

imputer = SimpleImputer(missing_values=np.nan, strategy='mean')

imputer.fit(df)

df = pd.DataFrame(data = imputer.transform(df), columns=df.columns)
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(df, y, test_size=.1, random_state=42)
```

```python
from sklearn.preprocessing import RobustScaler

sc = RobustScaler()
X_train = sc.fit_transform(X_train)
X_test = sc.transform(X_test)
```

```python
from sklearn.neighbors import KNeighborsClassifier
clf = KNeighborsClassifier(n_neighbors = 47)
clf.fit(X_train, y_train)
```

```python
from sklearn.metrics import confusion_matrix, accuracy_score
y_pred = clf.predict(X_test)
cm = confusion_matrix(y_test, y_pred)
print('Confusion matrix: ')
print(cm)
print('Accuracy: ', accuracy_score(y_test, y_pred)*100, '%')
```
