## Importing the libraries

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

## loading dataset

```python
df = pd.read_csv('/content/sonar data.csv')
X = df.iloc[:, :-1].values
y = df.iloc[:, -1].values
```

### Encoding the Dependent Variable

```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
y = le.fit_transform(y)
```

## Splitting the dataset into the Training set and Test set

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2, random_state = 1)
```

## Feature Scaling

```python
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
X_train = sc.fit_transform(X_train)
X_test = sc.transform(X_test)
```

### applying RandomForestClassifier

```python
from sklearn.ensemble import RandomForestClassifier
classifier = RandomForestClassifier(n_estimators = 55, criterion = 'entropy', random_state = 0)
classifier.fit(X_train, y_train)
```

### test set actual result and predicted result

```python
ts = pd.DataFrame(data=y_test, columns=['actual result'])
ts['predicted result'] = classifier.predict(X_test)
ts.head()
```

### confusion matrix and accuracy

```python
from sklearn.metrics import confusion_matrix, accuracy_score
y_pred = classifier.predict(X_test)
cm = confusion_matrix(y_test, y_pred)
print('confusion matrix:')
print(cm)
print('Accuracy: ', accuracy_score(y_test, y_pred)*100,'%')
```

### K-Fold cross validation & standard deviation

```python
from sklearn.model_selection import cross_val_score
accuracies = cross_val_score(estimator = classifier, X = X_train, y = y_train, cv = 10)
print("Accuracy: {:.2f} %".format(accuracies.mean()*100))
print("Standard Deviation: {:.2f} %".format(accuracies.std()*100))
```
